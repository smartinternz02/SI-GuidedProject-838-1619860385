# Matplot library

```python
import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)#
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'g')
plt.xlabel("linspace array")
plt.ylabel("powers of array")
plt.title("grapg b/w x and y")
plt.show()
```

```python
plt.scatter(x,y,color = "blue")
plt.xlabel("linspace array")
plt.ylabel("powers of array")
plt.title("graph b/w x and y")
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'g')
plt.subplot(2,2,4)
plt.plot(y,x,'r')
plt.subplot(2,2,2)
plt.scatter(x,y,color = "orange")
plt.subplot(2,2,3)
plt.scatter(y,x,color = "pink")
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'b')
axes2.scatter(x,y,color = 'red')
```

```python
plt.bar(x,y,color = "yellow")
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.scatterplot(x = "total_bill",y = "tip",data = tips)
```

```python
plt.scatter(tips["total_bill"],tips["tip"])
```

```python
sns.lineplot(x = "size",y = "tip",data = tips)
```

```python
sns.distplot(tips["tip"])
```

```python
sns.barplot(x = "time",y = "tip",data = tips)
```

```python
sns.boxplot(x = "size",y = "tip",data = tips)
```

```python
sns.catplot(x = "smoker",y = "tip",data = tips)
```

```python
sns.pairplot(tips)
```

```python
sns.heatmap(tips.corr(),annot = True)
```
